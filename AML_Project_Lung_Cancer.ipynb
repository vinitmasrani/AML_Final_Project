```python
import numpy as np
import cv2
import os
import numpy as np
from PIL import Image, ImageEnhance
```

```python
my_train_path = os.listdir(train_path)
```

```python
my_train_cancer = os.listdir(train_path + '/cancer/')
my_train_normal = os.listdir(train_path + '/normal/')
my_train_x = []
my_train_y = []
for image in my_train_cancer:
  img = cv2.imread(train_path + '/cancer/'+ image)
  image= Image.fromarray(img, 'RGB')
  image=image.resize((64, 64)) 
  my_train_x.append(np.array(image))
  my_train_y.append(0)
for image in my_train_normal:
  img = cv2.imread(train_path + '/normal/'+ image)
  image= Image.fromarray(img, 'RGB')
  image=image.resize((64, 64)) 
  my_train_x.append(np.array(image))
  my_train_y.append(1)
```

```python
print(len(my_train_x))
print(len(my_train_y))
print(my_train_y)
```

```python
my_test_cancer = os.listdir(test_path + '/cancer/')
my_test_normal = os.listdir(test_path + '/normal/')
my_test_x = []
my_test_y = []
for image in my_test_cancer:
  img = cv2.imread(test_path + '/cancer/'+ image)
  image= Image.fromarray(img, 'RGB')
  image=image.resize((64, 64)) 
  my_test_x.append(np.array(image))
  my_test_y.append(0)
for image in my_test_normal:
  img = cv2.imread(test_path + '/normal/'+ image)
  image= Image.fromarray(img, 'RGB')
  image=image.resize((64, 64)) 
  my_test_x.append(np.array(image))
  my_test_y.append(1)
```

```python
print(len(my_test_x))
print(len(my_test_y))
print(my_test_y)
```

```python
my_valid_cancer = os.listdir(valid_path + '/cancer/')
my_valid_normal = os.listdir(valid_path + '/normal/')
my_valid_x = []
my_valid_y = []
for image in my_valid_cancer:
  img = cv2.imread(valid_path + '/cancer/'+ image)
  image= Image.fromarray(img, 'RGB')
  image=image.resize((64, 64)) 
  my_valid_x.append(np.array(image))
  my_valid_y.append(0)
for image in my_valid_normal:
  img = cv2.imread(valid_path + '/normal/'+ image)
  image= Image.fromarray(img, 'RGB')
  image=image.resize((64, 64)) 
  my_valid_x.append(np.array(image))
  my_valid_y.append(1)
```

```python
print(len(my_valid_x))
print(len(my_valid_y))
print(my_valid_y)
```

```python
my_train_x = np.array(my_train_x)
my_train_y = np.array(my_train_y)
my_test_x = np.array(my_test_x)
my_test_y = np.array(my_test_y)
my_valid_x = np.array(my_valid_x)
my_valid_y = np.array(my_valid_y)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale = 1.0/255.0,
                                  horizontal_flip = True,
                                  fill_mode = 'nearest',
                                  zoom_range=0.2,
                                  shear_range = 0.2,
                                  width_shift_range=0.2,
                                  height_shift_range=0.2,
                                  rotation_range=0.4)

train_data = train_datagen.flow_from_directory(train_path,
                                               batch_size = 5,
                                               target_size = (64,64),
                                               class_mode = 'binary')
print(train_data.class_indices)
```

```python
from sklearn.utils import shuffle
test_datagen = ImageDataGenerator(rescale = 1.0/255.0)

test_data = test_datagen.flow_from_directory(test_path,
                                             batch_size = 5,
                                             target_size = (64,64),
                                             class_mode = 'binary')

print(test_data.class_indices)
```

```python
valid_datagen = ImageDataGenerator(rescale = 1.0/255.0)

valid_data = valid_datagen.flow_from_directory(valid_path,
                                               batch_size = 5,
                                               target_size = (64,64),
                                               class_mode = 'binary')
print(valid_data.class_indices)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.utils import normalize
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Activation, Dropout, Flatten, Dense
from tensorflow.keras.models import load_model, Model
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn import metrics
```

```python
model = Sequential()

model.add(Conv2D(32, (3,3), input_shape=(64, 64, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size =(2,2)))

model.add(Conv2D(64, (3,3),kernel_initializer='he_uniform'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size =(2,2)))

model.add(Conv2D(128, (3,3),kernel_initializer='he_uniform'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size =(2,2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy', optimizer= 'adam', metrics=['accuracy'])

print(model.summary())
```

```python
my_CNN_model = model.fit(my_train_x, my_train_y, 
                      batch_size = 16,
                      verbose = 1,
                      epochs = 32, 
                      validation_data = (my_valid_x,my_valid_y),
                      shuffle = False)
```

```python
print(my_CNN_model.history.keys())
predictions = model.predict(my_test_x)
```

```python
my_test_pred = []
for i in predictions:
  if(i<=0.5):
    my_test_pred.append(0)
  else:
    my_test_pred.append(1)
```

```python
my_test_pred = np.array(my_test_pred)
```

```python
def findAccuracy(data, prediction):
  count = 0
  true = 0
  datalen = len(data)
  for i in range(0, datalen):
    if data[i] != prediction[i]:
      count = count + 1
    else: 
      true = true + 1
  print(true)
```

```python
acc = findAccuracy(my_test_y, my_test_pred)
```
